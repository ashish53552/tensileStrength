```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv('test2.csv')
df.head()
```

```python
df.set_index(df['Unnamed: 0'])
df.drop(axis=1,columns=['Unnamed: 0'],inplace=True)
df
```

```python
df_bcc_phase=df[df['phases']=='BCC']
```

```python
X=df_bcc_phase[['R_delta', 'S_config', 'VEC',
       'R_cov_delta', 'density_avg', 'Tm_avg', 'E_avg', 'E_delta', 'G_avg',
       'G_delta', 'B_avg', 'B_delta', 'Compress_avg', 'Compress_delta',
       'EN_Allen_avg', 'H_avg', 'Vm_delta', 'E_coh_avg', 'E_coh_delta',
       'Senkov_param', 'H_ch_M_L_R', 'H_el_M_S_R']].values
y=df_bcc_phase[['YS_MPa']].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model=Sequential()
model.add(Dense(22,activation='softplus'))
model.add(Dense(22,activation='softplus'))
model.add(Dense(22,activation='softplus'))
# model.add(Dense(10,activation='softplus'))
# model.add(Dense(8,activation='softplus'))
# model.add(Dense(6,activation='softplus'))
# model.add(Dense(4,activation='softplus'))
# model.add(Dense(2,activation='softplus'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,epochs=2000)
```

```python
model.history.history
```

```python
loss=model.history.history['loss']
sns.lineplot(x=range(len(loss)),y=loss)
```
